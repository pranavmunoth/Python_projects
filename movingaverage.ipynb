```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ticker_symbol = "AAPL"
period = "1y"
stock = yf.Ticker(ticker_symbol)
data = stock.history(period=period)

```

```python
print(f"First 5 rows of {ticker_symbol} data")
data.head()
```

```python
print(f"/n Columns available")
print(data.columns.tolist())
```

```python
data["MA20"] = data["Close"].rolling(window=20).mean()
data["MA50"] = data["Close"].rolling(window=50).mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Close"], label = "Close Price", linewidth = 2)
plt.plot(data.index, data["MA20"], label = "MA20", linewidth = 1.5)
plt.plot(data.index, data["MA50"], label = "MA50", linewidth = 1.5)
plt.title(f"{ticker_symbol} - Close Price with Moving Average")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("Stock_Chart.png", dpi=300)
print("/nChart saved as Stock_Chart.png")
```
