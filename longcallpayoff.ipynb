```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def call_payoff(sT, strike_price, premium):
    pnl = np.where(sT>strike_price,sT-strike_price,0)
    return pnl-premium
```

```python
spot_price = 100
strike_price = 100
premium = 5
sT = np.arange(0.8*spot_price,1.2*spot_price)
```

```python
payoff_long_call = call_payoff(sT, strike_price, premium)
fig, ax = plt.subplots(figsize=(12,6))
ax.spines["bottom"].set_position("zero")
ax.plot(sT,payoff_long_call,label = "Long Call Payoff", linewidth = 2)
plt.grid(alpha=0.5)
plt.tight_layout()
plt.xlabel("Stock Price")
plt.ylabel("Profit and Loss")
plt.savefig("Long Call Payoff.png", dpi=300)
plt.legend()
plt.show()
```
